```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import locale
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' ) 
```

```python
df_gdp = pd.read_csv("./GDP/GDP_annual_growth_NEW.csv")
gdp = pd.DataFrame()
gdp['Country'] = df_gdp.pop('Country Name')
gdp['GDP Growth']    = df_gdp.pop('2020')
gdp.dropna(axis=0,inplace=True)
gdp
        
```

```python
dataframes = []
for i in range(1,5):
    df1 = pd.read_csv("./Ookla/fixed_year_2020_quarter_0{}.csv".format(i))
    df2 = pd.read_csv("./Ookla/fixed_year_2021_quarter_0{}.csv".format(i))
    df1['Avg. Avg U Kbps'] = df1['Avg. Avg U Kbps'].map(lambda x: locale.atof(x))
    df2['Avg. Avg U Kbps'] = df2['Avg. Avg U Kbps'].map(lambda x: locale.atof(x))
    df1['Avg. Avg D Kbps'] = df1['Avg. Avg D Kbps'].map(lambda x: locale.atof(x))
    df2['Avg. Avg D Kbps'] = df2['Avg. Avg D Kbps'].map(lambda x: locale.atof(x))
    df1['Avg Lat Ms'] = df1['Avg Lat Ms'].map(lambda x: float(x))
    df2['Avg Lat Ms'] = df2['Avg Lat Ms'].map(lambda x: float(x))
    dfout = pd.DataFrame()
    dfout['Mobile']   = np.zeros(np.shape(df1['Name']))
    dfout['Country']  = df1.pop('Name')
    dfout['Upload']   = np.subtract(df2.pop('Avg. Avg U Kbps'), df1.pop('Avg. Avg U Kbps'))
    dfout['Download'] = np.subtract(df2.pop('Avg. Avg D Kbps'), df1.pop('Avg. Avg D Kbps'))
    dfout['Latency']  = np.subtract(df2.pop('Avg Lat Ms'),df1.pop('Avg Lat Ms'))
    dfout.dropna(axis=0,inplace=True)
    dfout = dfout.merge(gdp, how='inner', on='Country')
    dataframes.append(dfout)

for i in range(1,5):
    df1 = pd.read_csv("./Ookla/mobile_year_2020_quarter_0{}.csv".format(i))
    df2 = pd.read_csv("./Ookla/mobile_year_2021_quarter_0{}.csv".format(i))
    df1['Avg. Avg U Kbps'] = df1['Avg. Avg U Kbps'].map(lambda x: locale.atof(x))
    df2['Avg. Avg U Kbps'] = df2['Avg. Avg U Kbps'].map(lambda x: locale.atof(x))
    df1['Avg. Avg D Kbps'] = df1['Avg. Avg D Kbps'].map(lambda x: locale.atof(x))
    df2['Avg. Avg D Kbps'] = df2['Avg. Avg D Kbps'].map(lambda x: locale.atof(x))
    df1['Avg Lat Ms'] = df1['Avg Lat Ms'].map(lambda x: float(x))
    df2['Avg Lat Ms'] = df2['Avg Lat Ms'].map(lambda x: float(x))
    dfout = pd.DataFrame()
    dfout['Mobile']   = np.ones(np.shape(df1['Name']))
    dfout['Country']  = df1.pop('Name')
    dfout['Upload']   = np.subtract(df2.pop('Avg. Avg U Kbps'), df1.pop('Avg. Avg U Kbps'))
    dfout['Download'] = np.subtract(df2.pop('Avg. Avg D Kbps'), df1.pop('Avg. Avg D Kbps'))
    dfout['Latency']  = np.subtract(df2.pop('Avg Lat Ms'),df1.pop('Avg Lat Ms'))
    dfout.dropna(axis=0,inplace=True)
    dfout = dfout.merge(gdp, how='inner', on='Country')
    dataframes.append(dfout)
    
```

```python
final_dataset = pd.concat(dataframes, axis=0)
final_dataset
```

```python
final_dataset.to_csv("./cleaned_dataset.csv", index=False)
```
